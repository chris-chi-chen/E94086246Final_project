```python
import tkinter as tk
import numpy as np
import math
import scipy
from matplotlib import pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
from PIL import Image,ImageTk
import scipy.linalg  
class basedesk():
    """Initial"""
    def __init__(self,master):
        self.root = master
        self.root.config()
        self.root.title('linear calculator')
        self.root.geometry('600x400')  
        self.root.resizable(0,0)##limit zoom
        homepage(self.root)##enter homepage                    
class homepage():
    def __init__(self,controller):     
        """create_page"""
        self.controller = controller
        self.initface = tk.Frame(self.controller, width=600, height=400)##create Frame
        self.initface.config(bg="white")##Color of Frame
        self.initface.pack(fill='both')
        # Read the Image
        image = Image.open("image1.png")
        # Resize the image using resize() method
        resize_image = image.resize((300, 400))
        img = ImageTk.PhotoImage(resize_image)
        # create label and add resize image
        label1 = tk.Label(self.initface,image=img)##put image
        label1.image = img##image
        label1.place(x=300,y=0)##arrange the location
        label=tk.Label(self.initface,text="Chris",bg='white',font=('Arial', 30))#create label of title name
        label.place(x=10,y=10)##arrange the location
        label2=tk.Label(self.initface,text="calculator",bg='white',font=('Arial', 30))#create label of title name
        label2.place(x=60,y=50)##arrange the location
        btn1 = tk.Button(self.initface,text='linear euqation',command=self.change)##button of change
        btn2 = tk.Button(self.initface,text='square matrix',command=self.change1)##button of change
        btn3 = tk.Button(self.initface,text='fitting',command=self.change2)##button of change
        btn1.place(x=80,y=300)##arrange the location
        btn2.place(x=270,y=300)##arrange the location
        btn3.place(x=470,y=300)##arrange the location
    def change(self):   
        """change page"""
        self.initface.destroy()#destroy
        face1(self.controller)##create pageThird     
    def change1(self):  
        """change page"""
        self.initface.destroy()#destroy
        face2(self.controller)##create pageSecond
    def change2(self):   
        """change page"""
        self.initface.destroy()#destroy
        face3(self.controller)##create pageThird  

class face1():
    def __init__(self,master):
        self.master = master #now root
        self.face1 = tk.Frame(self.master,width=600, height=400)#create frame
        self.var = tk.StringVar()#Variable of text of label  
        self.entries=[]#List of entry
        self.K=0 #vairable
        self.x=[] #variable x
        self.y=[] #variable y
        self.z=[] #variable z
        self.face1.config(bg="white")#color of frame
        self.face1.pack(fill='both')#fill entire zoom
        label1 = tk.Label(self.face1,bg='#3F704D',width=300,height=400)#background
        label1.place(x=300,y=0)#arrange the location
        btn_back = tk.Button(self.face1,text='back',command=self.back)#create back button
        btn_back.place(x=550,y=350)#arrange the location
        self.cal = tk.Frame(self.face1,width=400, height=200)#create frame of input
        self.cal.place(x=100,y=50)#arrange the location
        btn2 = tk.Button(self.face1,text='2_ele',command=self.two_lin)#create 2_ele button
        btn2.place(x=460,y=250)#arrange the location
        btn_3_3 = tk.Button(self.face1,text='3_ele',command=self.three_lin)#create 3_ele button
        btn_3_3.place(x=420,y=250)#arrange the location
        btn_4_4 = tk.Button(self.face1,text='4_ele',command=self.four_lin)#create 4_ele button
        btn_4_4.place(x=380,y=250)#arrange the location
        btn_ans = tk.Button(self.face1,text='ans',command=self.ans)#create get_answer button
        btn_ans.place(x=100,y=250)#arrange the location 
        self.ca3 = tk.Frame(self.face1,width=200, height=50)#create frame of showing answer
        self.ca3.place(x=130,y=320)#arrange the location 
        label2 = tk.Label(self.face1,   #answer label
                     text='answer:',  
                     bg="white",
                     font=('Arial', 15),
                     width=0, height=0)
        label2.place(x=130,y=290)#arrange the location
        label = tk.Label(self.ca3,       #answer of function
                     textvariable=self.var,  
                     font=('Arial', 12),
                     width=0, height=0)
        label.place(relx=0.5,rely=0.5,anchor='center')#arrange the location
    def two_lin(self):
        """create two elements of linear equation"""
        self.entries.clear()#clear list of entries
        for widgets in self.cal.winfo_children():##clear frame of input
              widgets.destroy()
        self.K=2#two element
        lax1=tk.Label(self.cal,text='x1',font=('Arial', 10))#x1 label
        lax1.place(x=60,y=60)#arrange the location 
        x=tk.Entry(self.cal,width=5)#Entry of x1
        x.insert(0, "0")#initial value is zero
        x.place(x=80,y=60)#arrange the location
        self.entries.append(x)#append in list of entries
        lax2=tk.Label(self.cal,text='x2',font=('Arial', 10))#x2 label
        lax2.place(x=120,y=60)#arrange the location 
        x=tk.Entry(self.cal,width=5)#Entry of x2
        x.insert(0, "0")#initial value is zero
        x.place(x=140,y=60)#arrange the location
        self.entries.append(x)#append in list of entries
        e1=tk.Label(self.cal,text='=',font=('Arial', 10))#equal label
        e1.place(x=180,y=60)#arrange the location
        x=tk.Entry(self.cal,width=5)#Entry of y1
        x.insert(0, "0")#initial value is zero
        x.place(x=200,y=60)#arrange the location
        self.entries.append(x)#append in list of entries
        lax3=tk.Label(self.cal,text='x1',font=('Arial', 10))#x1 label
        lax3.place(x=60,y=100)#arrange the location
        x=tk.Entry(self.cal,width=5)#Entry of x1
        x.insert(0, "0")#initial value is zero
        x.place(x=80,y=100)#arrange the location
        self.entries.append(x)#append in list of entries
        lax4=tk.Label(self.cal,text='x2',font=('Arial', 10))#x2 label
        lax4.place(x=120,y=100)#arrange the location 
        x=tk.Entry(self.cal,width=5)#Entry of x2
        x.insert(0, "0")#initial value is zero
        x.place(x=140,y=100)#arrange the location
        self.entries.append(x)#append in list of entries
        e2=tk.Label(self.cal,text='=',font=('Arial', 10))#equal label
        e2.place(x=180,y=100)#arrange the location
        x=tk.Entry(self.cal,width=5)#Entry of y2
        x.insert(0, "0")#initial value is zero
        x.place(x=200,y=100)#arrange the location
        self.entries.append(x)#append in list of entries
    def three_lin(self):
        """create three elements of linear equation"""
        self.entries.clear()#clear list of entries
        for widgets in self.cal.winfo_children():#clear frame of input
              widgets.destroy()
        self.K=3#three elements
        lax1=tk.Label(self.cal,text='x1',font=('Arial', 10))#x1 label
        lax1.place(x=60,y=60)#arrange the location 
        x=tk.Entry(self.cal,width=5)#Entry of x1
        x.insert(0, "0")#initial value is zero
        x.place(x=80,y=60)#arrange the location
        self.entries.append(x)#append in list of entries
        lax2=tk.Label(self.cal,text='x2',font=('Arial', 10))#x2 label
        lax2.place(x=120,y=60)#arrange the location
        x=tk.Entry(self.cal,width=5)#Entry of x2
        x.insert(0, "0")#initial value is zero
        x.place(x=140,y=60)#arrange the location
        self.entries.append(x)#append in list of entries
        lax3=tk.Label(self.cal,text='x3',font=('Arial', 10))#x3 label
        lax3.place(x=180,y=60)#arrange the location
        x=tk.Entry(self.cal,width=5)#Entry of x3
        x.insert(0, "0")#initial value is zero
        x.place(x=200,y=60)#arrange the location
        self.entries.append(x)#append in list of entries
        e1=tk.Label(self.cal,text='=',font=('Arial', 10))#equal label
        e1.place(x=240,y=60)#arrange the location
        x=tk.Entry(self.cal,width=5)#Entry of y1
        x.insert(0, "0")#initial value is zero
        x.place(x=260,y=60)#arrange the location
        self.entries.append(x)#append in list of entries
        lax1=tk.Label(self.cal,text='x1',font=('Arial', 10))#x1 label
        lax1.place(x=60,y=100)#arrange the location
        x=tk.Entry(self.cal,width=5)#Entry of x1
        x.insert(0, "0")#initial value is zero
        x.place(x=80,y=100)#arrange the location
        self.entries.append(x)#append in list of entries
        lax2=tk.Label(self.cal,text='x2',font=('Arial', 10))#x2 label
        lax2.place(x=120,y=100)#arrange the location
        x=tk.Entry(self.cal,width=5)#Entry of x2
        x.insert(0, "0")#initial value is zero
        x.place(x=140,y=100)#arrange the location
        self.entries.append(x)#append in list of entries
        lax3=tk.Label(self.cal,text='x3',font=('Arial', 10))#x3 label
        lax3.place(x=180,y=100)#arrange the location
        x=tk.Entry(self.cal,width=5)#Entry of x3
        x.insert(0, "0")#initial value is zero
        x.place(x=200,y=100)#arrange the location
        self.entries.append(x)#append in list of entries
        e1=tk.Label(self.cal,text='=',font=('Arial', 10))#equal label
        e1.place(x=240,y=100)#arrange the location
        x=tk.Entry(self.cal,width=5)#Entry of y2
        x.insert(0, "0")#initial value is zero
        x.place(x=260,y=100)#arrange the location
        self.entries.append(x)#append in list of entries
        lax1=tk.Label(self.cal,text='x1',font=('Arial', 10))#x1 label
        lax1.place(x=60,y=140)#arrange the location
        x=tk.Entry(self.cal,width=5)#Entry of x1
        x.insert(0, "0")#initial value is zero
        x.place(x=80,y=140)#arrange the location
        self.entries.append(x)#append in list of entries
        lax2=tk.Label(self.cal,text='x2',font=('Arial', 10))#x2 label
        lax2.place(x=120,y=140)#arrange the location
        x=tk.Entry(self.cal,width=5)#Entry of x2
        x.insert(0, "0")#initial value is zero
        x.place(x=140,y=140)#arrange the location
        self.entries.append(x)#append in list of entries
        lax3=tk.Label(self.cal,text='x3',font=('Arial', 10))#x3 label
        lax3.place(x=180,y=140)#arrange the location
        x=tk.Entry(self.cal,width=5)#Entry of x3
        x.insert(0, "0")#initial value is zero
        x.place(x=200,y=140)#arrange the location
        self.entries.append(x)#append in list of entries
        e1=tk.Label(self.cal,text='=',font=('Arial', 10))#equal label
        e1.place(x=240,y=140)#arrange the location
        x=tk.Entry(self.cal,width=5)#Entry of y3
        x.insert(0, "0")#initial value is zero
        x.place(x=260,y=140)#arrange the location
        self.entries.append(x)#append in list of entries
    def four_lin(self):
        """create four elements of linear equation"""
        self.entries.clear()#clear list of entries
        for widgets in self.cal.winfo_children():#clear frame of input
              widgets.destroy()
        self.K=4#four elements
        lax1=tk.Label(self.cal,text='x1',font=('Arial', 10))#x1 label
        lax1.place(x=60,y=40)#arrange the location
        x=tk.Entry(self.cal,width=5)#Entry of x1
        x.insert(0, "0")#initial value is zero
        x.place(x=80,y=40)#arrange the location
        self.entries.append(x)#append in list of entries
        lax2=tk.Label(self.cal,text='x2',font=('Arial', 10))#x2 label
        lax2.place(x=120,y=40)#arrange the location
        x=tk.Entry(self.cal,width=5)#Entry of x2
        x.insert(0, "0")#initial value is zero
        x.place(x=140,y=40)#arrange the location
        self.entries.append(x)#append in list of entries
        lax3=tk.Label(self.cal,text='x3',font=('Arial', 10))#x3 label
        lax3.place(x=180,y=40)#arrange the location
        x=tk.Entry(self.cal,width=5)#Entry of x3
        x.insert(0, "0")#initial value is zero
        x.place(x=200,y=40)#arrange the location
        self.entries.append(x)#append in list of entries
        lax4=tk.Label(self.cal,text='x4',font=('Arial', 10))#x4 label
        lax4.place(x=240,y=40)#arrange the location
        x=tk.Entry(self.cal,width=5)#Entry of x4
        x.insert(0, "0")#initial value is zero
        x.place(x=260,y=40)#arrange the location
        self.entries.append(x)#append in list of entries
        e1=tk.Label(self.cal,text='=',font=('Arial', 10))#equal label
        e1.place(x=300,y=40)#arrange the location
        x=tk.Entry(self.cal,width=5)#Entry of y1
        x.insert(0, "0")#initial value is zero
        x.place(x=320,y=40)#arrange the location
        self.entries.append(x)#append in list of entries
        lax1=tk.Label(self.cal,text='x1',font=('Arial', 10))#x1 label
        lax1.place(x=60,y=80)#arrange the location
        x=tk.Entry(self.cal,width=5)#Entry of x1
        x.insert(0, "0")#initial value is zero
        x.place(x=80,y=80)#arrange the location
        self.entries.append(x)#append in list of entries
        lax2=tk.Label(self.cal,text='x2',font=('Arial', 10))#x2 label
        lax2.place(x=120,y=80)#arrange the location
        x=tk.Entry(self.cal,width=5)#Entry of x2
        x.insert(0, "0")#initial value is zero
        x.place(x=140,y=80)#arrange the location
        self.entries.append(x)#append in list of entries
        lax3=tk.Label(self.cal,text='x3',font=('Arial', 10))#x3 label
        lax3.place(x=180,y=80)#arrange the location
        x=tk.Entry(self.cal,width=5)#Entry of x3
        x.insert(0, "0")#initial value is zero
        x.place(x=200,y=80)#arrange the location
        self.entries.append(x)#append in list of entries
        lax4=tk.Label(self.cal,text='x4',font=('Arial', 10))#x4 label
        lax4.place(x=240,y=80)#arrange the location
        x=tk.Entry(self.cal,width=5)#Entry of x4
        x.insert(0, "0")#initial value is zero
        x.place(x=260,y=80)#arrange the location
        self.entries.append(x)#append in list of entries
        e1=tk.Label(self.cal,text='=',font=('Arial', 10))#equal label
        e1.place(x=300,y=80)#arrange the location
        x=tk.Entry(self.cal,width=5)#Entry of y2
        x.insert(0, "0")#initial value is zero
        x.place(x=320,y=80)#arrange the location
        self.entries.append(x)#append in list of entries
        lax1=tk.Label(self.cal,text='x1',font=('Arial', 10))#x1 label
        lax1.place(x=60,y=120)#arrange the location
        x=tk.Entry(self.cal,width=5)#Entry of x1
        x.insert(0, "0")#initial value is zero
        x.place(x=80,y=120)#arrange the location
        self.entries.append(x)#append in list of entries
        lax2=tk.Label(self.cal,text='x2',font=('Arial', 10))#x2 label
        lax2.place(x=120,y=120)#arrange the location
        x=tk.Entry(self.cal,width=5)#Entry of x2
        x.insert(0, "0")#initial value is zero
        x.place(x=140,y=120)#arrange the location
        self.entries.append(x)#append in list of entries
        lax3=tk.Label(self.cal,text='x3',font=('Arial', 10))#x3 label
        lax3.place(x=180,y=120)#arrange the location
        x=tk.Entry(self.cal,width=5)#Entry of x3
        x.insert(0, "0")#initial value is zero
        x.place(x=200,y=120)#arrange the location
        self.entries.append(x)#append in list of entries
        lax4=tk.Label(self.cal,text='x4',font=('Arial', 10))#x4 label
        lax4.place(x=240,y=120)#arrange the location
        x=tk.Entry(self.cal,width=5)#Entry of x4
        x.insert(0, "0")#initial value is zero
        x.place(x=260,y=120)#arrange the location
        self.entries.append(x)#append in list of entries
        e1=tk.Label(self.cal,text='=',font=('Arial', 10))#equal label
        e1.place(x=300,y=120)#arrange the location
        x=tk.Entry(self.cal,width=5)#Entry of y3
        x.insert(0, "0")#initial value is zero
        x.place(x=320,y=120)#arrange the location
        self.entries.append(x)#append in list of entries
        lax1=tk.Label(self.cal,text='x1',font=('Arial', 10))#x1 label
        lax1.place(x=60,y=160)#arrange the location
        x=tk.Entry(self.cal,width=5)#Entry of x1
        x.insert(0, "0")#initial value is zero
        x.place(x=80,y=160)#arrange the location
        self.entries.append(x)#append in list of entries
        lax2=tk.Label(self.cal,text='x2',font=('Arial', 10))#x2 label
        lax2.place(x=120,y=160)#arrange the location
        x=tk.Entry(self.cal,width=5)#Entry of x2
        x.insert(0, "0")#initial value is zero
        x.place(x=140,y=160)#arrange the location
        self.entries.append(x)#append in list of entries
        lax3=tk.Label(self.cal,text='x3',font=('Arial', 10))#x3 label
        lax3.place(x=180,y=160)#arrange the location
        x=tk.Entry(self.cal,width=5)#Entry of x3
        x.insert(0, "0")#initial value is zero
        x.place(x=200,y=160)#arrange the location
        self.entries.append(x)#append in list of entries
        lax4=tk.Label(self.cal,text='x4',font=('Arial', 10))#x4 label
        lax4.place(x=240,y=160)#arrange the location
        x=tk.Entry(self.cal,width=5)#Entry of x4
        x.insert(0, "0")#initial value is zero
        x.place(x=260,y=160)#arrange the location
        self.entries.append(x)#append in list of entries
        e1=tk.Label(self.cal,text='=',font=('Arial', 10))#equal label
        e1.place(x=300,y=160)#arrange the location
        x=tk.Entry(self.cal,width=5)#Entry of y4
        x.insert(0, "0")#initial value is zero
        x.place(x=320,y=160)#arrange the location
        self.entries.append(x)#append in list of entries
    def ans(self):
        """get answer"""
        self.x.clear()#clear list
        self.y.clear()#clear list
        self.z.clear()#clear list
        NaN=0#wrong input detect
        i=1#count
        if(self.K==0):
             self.var.set(f"doesn't choose")#if choose or not
        else:
            for t in self.entries:#get entry 
                text=t.get()
                text.replace(' ','')
                if(text.lstrip( ' - ' ).isdigit() and text!=''):
                    self.z.append(int(text))#save all input
                    if(i%(self.K+1)!=0):
                        self.x.append(int(text))#save x
                    else:
                        self.y.append(int(text))#save y
                    i+=1
                else:
                    NaN=1#wrong!
                    break
            if(NaN==1):
                self.var.set(f"wrong input")##show wrong input
            else:
                A=np.resize(self.z,(self.K,self.K+1))#reszie change to np
                B=np.resize(self.x,(self.K,self.K))#resize change to np
                C=np.resize(self.y,(self.K,1))#reszie change to np
                rank1=np.linalg.matrix_rank(A)##get rank of all equal
                rank2=np.linalg.matrix_rank(B)##get rank of all equal except y
                if(rank1!=rank2):# if ranks aren't same
                     self.var.set(f"No solution") #wrong
                else:
                    if(rank1==0):# if ranks are zero 
                        self.var.set(f"All zero")
                    elif(rank1<self.K):# if ranks is small than variables
                        self.var.set(f"Infinitely many solutions")
                    else:
                        answer = np.linalg.solve(B, C)#have exact answer
                        if(self.K==2):
                            self.var.set(f"x1={np.around(answer[0],2)},x2={np.around(answer[1],2)}") 
                        elif(self.K==3):
                            self.var.set(f"x1={np.around(answer[0],2)},x2={np.around(answer[1],2)},x3={np.around(answer[2],2)}")
                        elif(self.K==4):
                            self.var.set(f"x1={np.around(answer[0],2)},x2={np.around(answer[1],2)}\nx3={np.around(answer[2],2)},x4={np.around(answer[3],2)}")                               
    def back(self):
        """back homepage"""
        self.face1.destroy() #destroy
        homepage(self.master)
class face2():
    def __init__(self,master):
        """create_page"""
        self.entries=[]#List of entry
        self.flag=0#check
        self.K=0#decide 2*2 3*3 4*4
        self.var = tk.StringVar()#variable of text  
        self.master = master
        self.matrix=[]#save matrix
        self.face2 = tk.Frame(self.master,width=600, height=400)#create frame
        self.face2.config(bg="white")#set coloer of frame
        self.face2.pack(fill='both')#fill entire zoom
        self.cal = tk.Frame(self.face2,width=400, height=200)#create show frame
        label1 = tk.Label(self.face2,bg='#3F704D',width=300,height=400)#background color
        label1.place(x=300,y=0)#arrange the location
        self.cal.place(x=100,y=50)#arrange the location
        btn_eigen=tk.Button(self.face2,text='eigenvalue',command=self.eigen)#button get eigen
        btn_eigen.place(x=100,y=250)#arrange the location
        btn_PLU=tk.Button(self.face2,text='PLU',command=self.LU)#button get PLU
        btn_PLU.place(x=100,y=280)#arrange the location
        btn_inver=tk.Button(self.face2,text='inverse',command=self.inver)#button get Inverse
        btn_inver.place(x=180,y=250) #arrange the location
        btn_det=tk.Button(self.face2,text='det',command=self.det)#button get det
        btn_det.place(x=240,y=250)#arrange the location
        btn_sq=tk.Button(self.face2,text='square',command=self.square)#button get square
        btn_sq.place(x=280,y=250)#arrange the location
        btn_submit = tk.Button(self.face2,text='submit',command=self.submit)#button save matrix
        btn_submit.place(x=340,y=250)#arrange the location
        btn_2_2 = tk.Button(self.face2,text='2*2',command=self.two)#button show 2*2 input
        btn_2_2.place(x=450,y=250)#arrange the location
        btn_3_3 = tk.Button(self.face2,text='3*3',command=self.three)#button show 3*3 input
        btn_3_3.place(x=420,y=250)#arrange the location
        btn_4_4 = tk.Button(self.face2,text='4*4',command=self.four)#button show 4*4 input
        btn_4_4.place(x=390,y=250)#arrange the location
        btn_back = tk.Button(self.face2,text='back',command=self.back)#button back
        btn_back.place(x=550,y=350)#arrange the location
        self.cal2 = tk.Frame(self.face2,bg="#e2e2e2",width=200, height=200)#create show answer frame
        self.cal2.place(x=300,y=50)#arrange the location
        label = tk.Label(self.cal2,#show answer
                     textvariable=self.var,  
                     bg="#e2e2e2",
                     font=('Arial', 8),
                     width=0, height=0)
        label.place(relx=0.5, rely=0.5,anchor='center')#arrange the location
        
    def back(self):
        """back homepage"""
        self.face2.destroy()
        homepage(self.master)
    def LU(self):
        """PLU"""
        if(self.flag == 0):#if input exists or not
            self.var.set(f"no matrix\ninput")
        else:#get PLU
            tM=np.resize( np.array(self.matrix), (int(math.sqrt(self.K)),int(math.sqrt(self.K))))
            P,L,U=scipy.linalg.lu(tM)
            self.var.set(f"P is :{np.around(P, 2)}\nL is :{np.around(L, 2)}\nU is :{np.around(U, 2)}")
    def square(self):
        """square"""
        if(self.flag == 0):#if input exists or not
            self.var.set(f"no matrix\ninput")
        else:#get Square
            tM=np.resize( np.array(self.matrix), (int(math.sqrt(self.K)),int(math.sqrt(self.K))))
            self.var.set(f"Square are :\n{np.around(np.dot(tM,tM), 2)}")
    def eigen(self):
        """eigen"""
        if(self.flag == 0):#if input exists or not
            self.var.set(f"no matrix\ninput")
        else:#get eigenvalue
            tM=np.resize( np.array(self.matrix), (int(math.sqrt(self.K)),int(math.sqrt(self.K))))
            value ,vector = np.linalg.eig(tM)   
            self.var.set(f"Eigenvalues are :\n{np.around(value, 2)}") 
    def inver(self):
        """inverse"""
        if(self.flag == 0):#if input exists or not
            self.var.set(f"no matrix\ninput")
        else:#get inverse
            tM=np.resize( np.array(self.matrix), (int(math.sqrt(self.K)),int(math.sqrt(self.K)))) 
            if(abs(np.linalg.det(tM))>=1):#det!=0           
                self.var.set(f"The inverse:\n{np.around(np.linalg.inv(tM), 2)}")
            else:#det=0
                self.var.set(f"determinant is \nzero")
    def det(self):
        if(self.flag == 0):#if input exists or not
            self.var.set(f"no matrix\ninput")
        else:#get det
            tM=np.resize( np.array(self.matrix), (int(math.sqrt(self.K)),int(math.sqrt(self.K))))   
            dett=np.linalg.det(tM)
            if(abs(dett)<1):#det=0
                self.var.set(f"The det:\n0")
            else:#det!=0 
                self.var.set(f"The det:\n{np.around(dett, 2)}")
    def submit(self):
        self.matrix.clear()#clear matrix
        self.flag=0#flag=0
        if(self.K==0):#not choose
             self.var.set(f"do not meet \nthe requirements")
        else:
            for t in self.entries:#get entry
                text=t.get()
                text.replace(' ','')
                if(text.lstrip( ' - ' ).isdigit() and text!=''):##limit positive negative integer
                    self.matrix.append(int(text))
            if len(self.matrix)==self.K:##if rationable or not
                tM=np.resize( np.array(self.matrix), (int(math.sqrt(self.K)),int(math.sqrt(self.K))))
                self.var.set(f"now matrix:\n{tM}") 
                self.flag=1
            else:
                self.var.set(f"do not meet \nthe requirements")   
    def two(self):
        self.entries.clear()##clear list of entries
        self.K=0
        for widgets in self.cal.winfo_children():#clear show frame
              widgets.destroy()
        self.K=4#2*2
        x=tk.Entry(self.cal,width=5)#x1
        x.place(x=50,y=50)#arrange the location
        self.entries.append(x)
        x=tk.Entry(self.cal,width=5)#x2
        x.place(x=100,y=50)#arrange the location
        self.entries.append(x)
        x=tk.Entry(self.cal,width=5)#x3
        x.place(x=50,y=100)#arrange the location
        self.entries.append(x)
        x=tk.Entry(self.cal,width=5)#x4
        x.place(x=100,y=100)#arrange the location
        self.entries.append(x)
    def three(self):
        self.entries.clear()##clear list of entries
        self.K=0
        for widgets in self.cal.winfo_children():#clear show frame
              widgets.destroy()
        self.K=9#3*3
        x=tk.Entry(self.cal,width=5)#x1
        x.place(x=25,y=50)#arrange the location
        self.entries.append(x)
        x=tk.Entry(self.cal,width=5)#x2
        x.place(x=75,y=50)#arrange the location
        self.entries.append(x)
        x=tk.Entry(self.cal,width=5)#x3
        x.place(x=125,y=50)#arrange the location
        self.entries.append(x)
        x=tk.Entry(self.cal,width=5)#x4
        x.place(x=25,y=75)#arrange the location
        self.entries.append(x)
        x=tk.Entry(self.cal,width=5)#x5
        x.place(x=75,y=75)#arrange the location
        self.entries.append(x)
        x=tk.Entry(self.cal,width=5)#x6
        x.place(x=125,y=75)#arrange the location
        self.entries.append(x)
        x=tk.Entry(self.cal,width=5)#x7
        x.place(x=25,y=100)#arrange the location
        self.entries.append(x)
        x=tk.Entry(self.cal,width=5)#x8
        x.place(x=75,y=100)#arrange the location
        self.entries.append(x)
        x=tk.Entry(self.cal,width=5)#x9
        x.place(x=125,y=100)#arrange the location
        self.entries.append(x)
    def four(self):
        self.entries.clear()##clear list of entries
        self.K=0
        for widgets in self.cal.winfo_children():#clear show frame
              widgets.destroy()
        self.K=16#4*4
        x=tk.Entry(self.cal,width=5)#x1
        x.place(x=10.5,y=40)#arrange the location
        self.entries.append(x)
        x=tk.Entry(self.cal,width=5)#x2
        x.place(x=55.5,y=40)#arrange the location
        self.entries.append(x)
        x=tk.Entry(self.cal,width=5)#x3
        x.place(x=105.5,y=40)#arrange the location
        self.entries.append(x)
        x=tk.Entry(self.cal,width=5)#x4
        x.place(x=155.5,y=40)#arrange the location
        self.entries.append(x) 
        x=tk.Entry(self.cal,width=5)#x5
        x.place(x=10.5,y=65)#arrange the location
        self.entries.append(x)
        x=tk.Entry(self.cal,width=5)#x6
        x.place(x=55.5,y=65)#arrange the location
        self.entries.append(x)
        x=tk.Entry(self.cal,width=5)#x7
        x.place(x=105.5,y=65)#arrange the location
        self.entries.append(x)
        x=tk.Entry(self.cal,width=5)#x8
        x.place(x=155.5,y=65)#arrange the location
        self.entries.append(x)   
        x=tk.Entry(self.cal,width=5)#x9
        x.place(x=10.5,y=90)#arrange the location
        self.entries.append(x)
        x=tk.Entry(self.cal,width=5)#x10
        x.place(x=55.5,y=90)#arrange the location
        self.entries.append(x)
        x=tk.Entry(self.cal,width=5)#x11
        x.place(x=105.5,y=90)#arrange the location
        self.entries.append(x)
        x=tk.Entry(self.cal,width=5)#x12
        x.place(x=155.5,y=90)#arrange the location
        self.entries.append(x)      
        x=tk.Entry(self.cal,width=5)#x13
        x.place(x=10.5,y=115)#arrange the location
        self.entries.append(x)
        x=tk.Entry(self.cal,width=5)#x14
        x.place(x=55.5,y=115)#arrange the location
        self.entries.append(x)
        x=tk.Entry(self.cal,width=5)#x15
        x.place(x=105.5,y=115)#arrange the location
        self.entries.append(x)
        x=tk.Entry(self.cal,width=5)#x16
        x.place(x=155.5,y=115)#arrange the location
        self.entries.append(x)    
class face3():
    def __init__(self,master):
        """create page"""
        self.master = master
        self.var = tk.StringVar()#Variable of text of label
        self.face3 = tk.Frame(self.master,width=600, height=400)#create frame
        self.face3.config(bg="white")#color of frame
        self.face3.pack(fill='both')#fill entire zoom
        label1 = tk.Label(self.face3,bg='#3F704D',width=300,height=400)#create backgroun
        self.cal = tk.Frame(self.face3,bg='#3F704D',width=300, height=400)#create show frame
        label1.place(x=300,y=0)#arrange the location
        self.cal.place(x=300,y=0)
        btn_back = tk.Button(self.face3,text='back',command=self.back)#back button
        btn_back.place(x=550,y=350) #arrange the location
        btn_d1 = tk.Button(self.face3,text='degree1',command=self.d1)#get ax+b
        btn_d1.place(x=20,y=365) #arrange the location
        btn_d1 = tk.Button(self.face3,text='degree2',command=self.d2)#get ax^2+bx+c
        btn_d1.place(x=90,y=365) #arrange the location
        btn_d1 = tk.Button(self.face3,text='degree3',command=self.d3)#get ax^3+bx^2+cx+d
        btn_d1.place(x=160,y=365) #arrange the location
        self.entriesx=[]#list of extries of x
        self.entriesy=[]#list of extries of y
        label_x=tk.Label(self.face3,text='x',font=('Arial', 15),bg='white')#x
        label_x.place(x=10,y=300)#arrange the location
        label_y=tk.Label(self.face3,text='y',font=('Arial', 15),bg='white')#y
        label_y.place(x=10,y=330)#arrange the location
        label=tk.Label(self.face3,text='1',font=('Arial', 13),bg='white')#1
        label.place(x=60,y=270)#arrange the location
        x=tk.Entry(self.face3,width=3,bg='#FFEF9F')#x1
        x.place(x=55,y=305)#arrange the location
        self.entriesx.append(x)
        y=tk.Entry(self.face3,width=3,bg='#FFEF9F')#y1
        y.place(x=55,y=335)#arrange the location
        self.entriesy.append(y)
        label=tk.Label(self.face3,text='2',font=('Arial', 13),bg='white')#2
        label.place(x=100,y=270)#arrange the location
        x=tk.Entry(self.face3,width=3,bg='#FFEF9F')#x2
        x.place(x=95,y=305)#arrange the location
        self.entriesx.append(x)
        y=tk.Entry(self.face3,width=3,bg='#FFEF9F')#y2
        y.place(x=95,y=335)#arrange the location
        self.entriesy.append(y)
        label=tk.Label(self.face3,text='3',font=('Arial', 13),bg='white')#3
        label.place(x=140,y=270)#arrange the location
        x=tk.Entry(self.face3,width=3,bg='#FFEF9F')#x3
        x.place(x=135,y=305)#arrange the location
        self.entriesx.append(x)
        y=tk.Entry(self.face3,width=3,bg='#FFEF9F')#y3
        y.place(x=135,y=335)#arrange the location
        self.entriesy.append(y)
        label=tk.Label(self.face3,text='4',font=('Arial', 13),bg='white')#4
        label.place(x=180,y=270)#arrange the location
        x=tk.Entry(self.face3,width=3,bg='#FFEF9F')#x4
        x.place(x=175,y=305)#arrange the location
        self.entriesx.append(x)
        y=tk.Entry(self.face3,width=3,bg='#FFEF9F')#y4
        y.place(x=175,y=335)#arrange the location
        self.entriesy.append(y)
        label=tk.Label(self.face3,text='5',font=('Arial', 13),bg='white')#5
        label.place(x=220,y=270)#arrange the location
        x=tk.Entry(self.face3,width=3,bg='#FFEF9F')#x5
        x.place(x=215,y=305)#arrange the location
        self.entriesx.append(x)
        y=tk.Entry(self.face3,width=3,bg='#FFEF9F')#y5
        y.place(x=215,y=335)#arrange the location
        self.entriesy.append(y)
        label=tk.Label(self.face3,text='6',font=('Arial', 13),bg='white')#6
        label.place(x=260,y=270)#arrange the location
        x=tk.Entry(self.face3,width=3,bg='#FFEF9F')#x6
        x.place(x=255,y=305)#arrange the location
        self.entriesx.append(x)
        y=tk.Entry(self.face3,width=3,bg='#FFEF9F')#y6
        y.place(x=255,y=335)#arrange the location
        self.entriesy.append(y)
        label=tk.Label(self.face3,text='7',font=('Arial', 13),bg='#3F704D',fg = 'white')#7
        label.place(x=300,y=270)#arrange the location
        x=tk.Entry(self.face3,width=3,bg='#FFEF9F')#x7
        x.place(x=295,y=305)#arrange the location
        self.entriesx.append(x)
        y=tk.Entry(self.face3,width=3,bg='#FFEF9F')#y7
        y.place(x=295,y=335)#arrange the location
        self.entriesy.append(y)
        label=tk.Label(self.face3,text='8',font=('Arial', 13),bg='#3F704D',fg = 'white')#8
        label.place(x=340,y=270)#arrange the location
        x=tk.Entry(self.face3,width=3,bg='#FFEF9F')#x8
        x.place(x=335,y=305)#arrange the location
        self.entriesx.append(x)
        y=tk.Entry(self.face3,width=3,bg='#FFEF9F')#y8
        y.place(x=335,y=335)#arrange the location
        self.entriesy.append(y)
        label=tk.Label(self.face3,text='9',font=('Arial', 13),bg='#3F704D',fg = 'white')#9
        label.place(x=380,y=270)#arrange the location
        x=tk.Entry(self.face3,width=3,bg='#FFEF9F')#x9
        x.place(x=375,y=305)#arrange the location
        self.entriesx.append(x)
        y=tk.Entry(self.face3,width=3,bg='#FFEF9F')#y9
        y.place(x=375,y=335)#arrange the location
        self.entriesy.append(y)
        label=tk.Label(self.face3,text='10',font=('Arial', 13),bg='#3F704D',fg = 'white')#10
        label.place(x=415,y=270)#arrange the location
        x=tk.Entry(self.face3,width=3,bg='#FFEF9F')#x10
        x.place(x=415,y=305)#arrange the location
        self.entriesx.append(x)
        y=tk.Entry(self.face3,width=3,bg='#FFEF9F')#y10
        y.place(x=415,y=335)#arrange the location
        self.entriesy.append(y)
        label = tk.Label(self.face3,#show answer
                     textvariable=self.var,  
                     bg="white",
                     font=('Arial', 15),
                     width=0, height=0)
        label.place(x=85, y=30)#arrange the location
    def d1(self):
        """y=ax+b"""
        temp_x=[]#create list
        temp_y=[]#create list
        x_l=[]#create list
        y_l=[]#create list
        for t in self.entriesx: #save entries of x 
                text=t.get()
                text.replace(' ','')
                temp_x.append(text)
        for t in self.entriesy: #save entries of y
                text=t.get()
                text.replace(' ','')
                temp_y.append(text)
        for i in range(0,10):#if input is rationable or not
            if(temp_x[i].lstrip( ' - ' ).isdigit() and temp_y[i].lstrip( ' - ' ).isdigit()):
                x_l.append(int(temp_x[i]))
                y_l.append(int(temp_y[i]))
        length=len(x_l)
        if(length<2):## if rationable input < 2 sets
            self.var.set(f"not meet")
        else:
            tx=np.array(x_l)#get np array of x
            ty=np.array(y_l)#get np array of y
            coefficients = np.polyfit(tx, ty, 1)#fitting
            figure=plt.figure(figsize=(2.5,2.5),dpi=100)#create plot
            figure.add_subplot(111).plot(x_l,y_l,'o')#draw data
            x_l=sorted(x_l)
            plt.plot(np.array(x_l),np.polyval(coefficients ,np.array(x_l)))#draw y=ax+b
            chart=FigureCanvasTkAgg(figure,self.cal)#make matplot able to put in tk
            chart.get_tk_widget().place(relx=0.5, rely=0.35,anchor='center')#arrange the location
            self.var.set(f"ax+b\na={np.around(coefficients[0],2)}\nb={np.around(coefficients[1],2)}")#show answer
    def d2(self):
        """y=ax^2+bx+c"""
        temp_x=[]#create list
        temp_y=[]#create list
        x_l=[]#create list
        y_l=[]#create list
        for t in self.entriesx: #save entries of x 
                text=t.get()
                text.replace(' ','')
                temp_x.append(text)
        for t in self.entriesy: #save entries of y
                text=t.get()
                text.replace(' ','')
                temp_y.append(text)
        for i in range(0,10):#if input is rationable or not
            if(temp_x[i].lstrip( ' - ' ).isdigit() and temp_y[i].lstrip( ' - ' ).isdigit()):
                x_l.append(int(temp_x[i]))
                y_l.append(int(temp_y[i]))
        length=len(x_l)
        if(length<3):## if rationable input < 3 sets
            self.var.set(f"not meet")
        else:
            tx=np.array(x_l)#get np array of x
            ty=np.array(y_l)#get np array of y
            coefficients = np.polyfit(tx, ty, 2)#fitting
            figure=plt.figure(figsize=(2.5,2.5),dpi=100)#create plot
            figure.add_subplot(111).plot(x_l,y_l,'o')#draw data
            x_l=sorted(x_l)
            plt.plot(np.array(x_l),np.polyval(coefficients ,np.array(x_l)))#draw y=ax^2+bx+c
            chart=FigureCanvasTkAgg(figure,self.cal)#make matplot able to put in tk
            chart.get_tk_widget().place(relx=0.5, rely=0.35,anchor='center')#arrange the location
            self.var.set(f"ax^2+bx+c\na={np.around(coefficients[0],2)}\nb={np.around(coefficients[1],2)}\nc={np.around(coefficients[2],2)}")#show answer
    def d3(self):
        """y=ax^3+bx^2+cx+d"""
        temp_x=[]#create list
        temp_y=[]#create list
        x_l=[]#create list
        y_l=[]#create list
        for t in self.entriesx: #save entries of x 
                text=t.get()
                text.replace(' ','')
                temp_x.append(text)
        for t in self.entriesy: #save entries of y
                text=t.get()
                text.replace(' ','')
                temp_y.append(text)
        for i in range(0,10):#if input is rationable or not
            if(temp_x[i].lstrip( ' - ' ).isdigit() and temp_y[i].lstrip( ' - ' ).isdigit()):
                x_l.append(int(temp_x[i]))
                y_l.append(int(temp_y[i]))
        length=len(x_l)
        if(length<4):## if rationable input < 4 sets
            self.var.set(f"not meet")
        else:
            tx=np.array(x_l)#get np array of x
            ty=np.array(y_l)#get np array of y
            coefficients = np.polyfit(tx, ty, 3)#fitting
            figure=plt.figure(figsize=(2.5,2.5),dpi=100)#create plot
            figure.add_subplot(111).plot(x_l,y_l,'o')#draw data
            x_l=sorted(x_l)
            plt.plot(np.array(x_l),np.polyval(coefficients ,np.array(x_l)))#draw y=ax^3+bx^2+cx+d
            chart=FigureCanvasTkAgg(figure,self.cal)#make matplot able to put in tk
            chart.get_tk_widget().place(relx=0.5, rely=0.35,anchor='center')#arrange the location
            self.var.set(f"ax^3+bx^2+cx+d\na={np.around(coefficients[0],2)}\nb={np.around(coefficients[1],2)}\nc={np.around(coefficients[2],2)}\nd={np.around(coefficients[3],2)}") #show answer
    def back(self):
        """back homepage"""
        self.face3.destroy()
        homepage(self.master)
if __name__ == '__main__':    
    root = tk.Tk()#create root
    basedesk(root)
    root.mainloop()
```
